```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

```python
url= 'https://wuzzuf.net/search/jobs/?a=navbl%7Cspbl&q=data%20analyst'
```

```python
html= urlopen(url).read()
```

```python
soup= bs(html ,'html.parser')
```

```python
jobs= soup.find_all('div', class_='css-ghe2tq e1v1l3u10')
```

```python
print(len(jobs))
```

```python
with open ('wuzzef_data_analyst.csv','w',encoding ='utf-8', newline='') as file:
    file.write('job_title, company_name, job_type, work_mode\n')
    for job in jobs:
        job_title= job.find('h2', class_='css-193uk2c')
        company_name= job.find('a', class_='css-ipsyv7')
        job_type= job.find('span', class_='css-uc9rga eoyjyou0')
        work_mode= job.find('span', class_='css-uofntu eoyjyou0')

        file.write(
             f'{job_title.text.strip() if job_title else 'N/A' },'
             f'{company_name.text.strip() if company_name else 'N/A' },'
             f'{job_type.text.strip() if job_type else 'N/A' },'
             f'{work_mode.text.strip() if work_mode else 'N/A' }\n') 
```
